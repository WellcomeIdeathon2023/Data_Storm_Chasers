```python
#Just a shell for the user interface. Supposed to be able to select the data and features required 
# for the code in 'computation' file.

from sklearn import datasets
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import re
from bokeh.plotting import figure, show, output_notebook
import ipywidgets as widgets
from IPython.display import display, clear_output

app_layout = widgets.Layout(display='flex',
                flex_flow='row nowrap',
                align_items='center',
                border='none',
                width='100%',
                margin='5px 5px 5px 5px')

main_display = widgets.Output(layout=app_layout)

#def show_app(inputs):
#    with main_display:
#        main_display.clear_output()
#        display(inputs)

#upload = widgets.FileUpload(
#    accept='.*',
#    multiple=True,
#    #description='Upload Data'
#)

#upload.observe(show_app, names='value')

trade_data = widgets.Dropdown(
    options=[('Bilateral Trade', 'data/trade/Bilateral_Trade_in_Low_Carbon_Technology_Products.csv'),
             ('Trade', 'data/trade/Trade_in_Low_Carbon_Technology_Products.csv')],
    #value=0,
    description='Trade data:',
)

econ_data = widgets.Dropdown(
    options=[('Direct Investment', 'data/economy/Direct_Investment-related_Indicators_IMF.csv'),
             ('Environmental Protection', 'data/economy/Environmental_Protection_Expenditures_IMF.csv'),
             ('Environmental Taxes','data/economy/Environmental_Taxes_IMF.csv'),
             ('Fossil Fuel Subsidies','data/economy/Fossil_Fuel_Subsidies_IMF.csv')],
    #value=1,
    description='Economy data:',
)

climate_data = widgets.Dropdown(
    options=[('Temperature Change', 'data/climate/Annual_Surface_Temperature_Change.csv'),
             ('Forest and Carbon', 'data/climate/Forest_and_Carbon.csv')],
    #value=1,
    description='Climate data:',
)

data_lists = widgets.GridspecLayout(4,3)
data_lists[0,:] = widgets.HBox([trade_data, econ_data, climate_data])
display(data_lists)
```

```python
#show_app([trade_data, econ_data, climate_data])
#display(trade_data)
def list_choice(data_list):
    print('You chose {}'.format(data_list))
    
list_choices_Trade = widgets.interactive_output(list_choice, {'data_list':trade_data})
#list_choices_Econ = widgets.interactive_output(list_choice, {'data_list':econ_data, 'descr':'Economic Data'})
#list_choices_Climate = widgets.interactive_output(list_choice, {'data_list':climate_data, 'descr':'Climate Data'})

def read_file_param(file_path, country_names='ISO3', unit_column = 'Unit'):
    fl = pd.read_csv(file_path)
    columns = fl.columns
    no_columns = len(columns)
    ISO3_index = columns.get_loc(country_names)
    unit_index = columns.get_loc(unit_column)
    feature_names = np.unique(fl[[unit_column]].values)
    years_list=[]
    for index, header in enumerate(columns):
        if re.match('F[1-2][0-9][0-9][0-9]', header):
            years_list.append((index,header[1:]))
    years = [years_list[0][0], years_list[-1][0], years_list[0][1], years_list[-1][1]]
    #print(no_columns,'\n', list(columns), '\n', ISO3_index, unit_index, years)
    return [columns, no_columns, ISO3_index, unit_index, list(feature_names), years]

def read_unique_values(file_path, column_name):
    fl = pd.read_csv(file_path)
    if type(column_name)==int:
        str_column_name = list(fl.columns)[column_name]
    else:
        str_column_name = column_name
    feature_name = np.unique(fl[[str_column_name]].values)
    return list(feature_name)



trade_file_param = read_file_param(trade_data.value)
econ_file_param = read_file_param(econ_data.value)
climate_file_param = read_file_param(climate_data.value)

trade_data.disabled=True
econ_data.disabled = True
climate_data.disabled = True



trade_file_feature1 = widgets.Dropdown(
    options=trade_file_param[0][1:],#read_unique_values(trade_data.value, 'Indicator'),
    #value=1,
    description='Feature1 for trade data',
)

trade_file_feature2 = widgets.Dropdown(
    options=trade_file_param[0][1:],#read_unique_values(trade_data.value, 8),
#    #value=1,
    description='Feature2 for trade data',
)


econ_file_feature1 = widgets.Dropdown(
    options=econ_file_param[0][1:],
    #value=1,
    description='Feature1 name for economic data',
)

econ_file_feature2 = widgets.Dropdown(
    options=econ_file_param[0][1:],#read_unique_values(econ_data.value, 6),
    #value=1,
    description='Feature2 name for economic data',
)

climate_file_feature1 = widgets.Dropdown(
    options=climate_file_param[0][1:],#read_unique_values(climate_data.value, 4),
    #value=1,
    description='Feature1 name for climate data',
)

climate_file_feature2 = widgets.Dropdown(
    options=climate_file_param[0][1:],#read_unique_values(climate_data.value, 9),
    #value=1,
    description='Feature2 name for climate data',
)

years_trade = widgets.IntRangeSlider(
    min=1950,
    max=2050,
    step=1,
    description='Chosen Years for Trade',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='d',
)
#year_start_trade.value
years_econ = widgets.IntRangeSlider(
    min=1950,
    max=2050,
    step=1,
    description='Chosen Years for Economy',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='d',
)
years_climate = widgets.IntRangeSlider(
    min=1950,
    max=2050,
    step=1,
    description='Chosen Years for Climate',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='d',
)

#data_lists = widgets.VBox(
#    [widgets.HBox([trade_data, econ_data, climate_data]),
#     widgets.HBox([trade_file_feature1, econ_file_feature_name, climate_file_feature_name]),
#     widgets.HBox([trade_file_feature2]),
#    widgets.HBox([years_trade, years_econ, years_climate])
#     ])

data_lists = widgets.GridspecLayout(4,3)
data_lists[0,:] = widgets.HBox([trade_data, econ_data, climate_data])
data_lists[1,:] = widgets.HBox([trade_file_feature1, econ_file_feature1, climate_file_feature1])
data_lists[2,:] = widgets.HBox([trade_file_feature2, econ_file_feature2, climate_file_feature2])
data_lists[3,:] = widgets.HBox([years_trade, years_econ, years_climate])

main_display.clear_output()
display(data_lists)
```
