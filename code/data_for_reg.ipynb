```python
trade_drop=['ObjectId', 'Country', 'ISO2', 
            #'ISO3', 
            'Counterpart_Country','Counterpart_ISO2', 
            #'Counterpart_ISO3', 
            'Indicator', 'Unit', 
            'Source',
       'CTS_Code', 'CTS_Name', 'CTS_Full_Descriptor', 'Trade_Flow', 'Scale',
      # 'F1994', 'F1995', 'F1996', 'F1997', 'F1998', 'F1999', 'F2000', 'F2001',
       #'F2002', 'F2003', 'F2004', 'F2005', 'F2006', 'F2007', 'F2008', 'F2009',
       #'F2010', 'F2011', 'F2012', 'F2013', 'F2014', 'F2015', 'F2016', 'F2017',
       #'F2018', 'F2019', 'F2020', 'F2021'
           ]
df_trade=df_trade.drop(trade_drop,axis=1)
df_trade
```

```python
new_column_names = {
    'Expenditure on environment protection': 'Environment_Expenditure',
    'CO2 emissions per unit of output for foreign controlled multinational enterprises': 'CO2 emissions_Foreign_control',
    'CO2 emissions per unit of output of domestic controlled enterprises': 'CO2_emissions_Domestic_control',
    'Environmental Taxes': 'Taxes_Environment',
    'Fossil Fuel Subsidies - Total Implicit and Explicit': 'Fossil_Fuel_Subsidies',
    'G7': 'G7_dummy',
    'Temperature change with respect to a baseline climatology, corresponding to the period 1951-1980': 'Temperature_Change'
}

combined_df.rename(columns=new_column_names, inplace=True)
combined_df
```

The following runs a Panel OLS regression of the form:
$$y_{it}=\beta_0 + \beta_1x_{it}+ e_{it}$$

```python
model1 = PanelOLS.from_formula("Environment_Expenditure ~ 1+Temperature_Change ", data=combined_df)
print(model1.fit())
```

```python
model2 = PanelOLS.from_formula("Taxes_Environment ~ 1+Temperature_Change ", data=combined_df)
print(model2.fit())
```

The following runs a Panel OLS with *random effects* regression of the form:
$$y_{it}=\beta_0 + \beta_1x_{it}+ country_i+e_{it}$$

```python
from linearmodels.panel import RandomEffects
model3 = RandomEffects.from_formula("Environment_Expenditure ~ 1+Temperature_Change +EntityEffects", data=combined_df)
print(model3.fit())
```

The following runs a Panel OLS with *random effects* regression of the form:
$$y_{it}=\beta_0 + \beta_1x_{it} +\beta_1 x_{it} G7+ country_i+e_{it}$$

```python
model4 = RandomEffects.from_formula("Environment_Expenditure ~ 1 + Temperature_Change+G7_dummy*Temperature_Change +EntityEffects", data=combined_df)
print(model4.fit())
```
